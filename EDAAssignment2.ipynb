Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type

```python
df.head(2)
```

```python
df=df.drop(df.index[10472])
```

```python
df['Price'].unique()
```

```python
chars_to_remove=['+',',','$']
cols_to_clean=['Price']
for item in chars_to_remove:
    for cols in cols_to_clean:
         df[cols]=df[cols].str.replace(item,'')
```

```python
df['Price']=df['Price'].astype('float')
```

```python
df['Type'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
```

```python
df['Type']=encoder.fit_transform(df['Type'])
```

```python
df.info()
```

```python
df['Type'].unique()
```

```python
avg_price_by_type=df.groupby('Type')['Price'].mean()
avg_price_by_type.plot(kind='bar')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.title('Average Price by App Type')
plt.show()
```

As you can see from the bar chart, paid apps have a higher average price than free apps. This is to be expected there are some free apps that have a non-zero price, which may be due to in-app purchases or other monetization.

Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

```python
install_freq = df[['App', 'Installs']].groupby('App').sum()
```

```python
install_freq
```

```python
# Sort the frequency table in descending order and select the top 10 apps
top_10_apps = install_freq.sort_values('Installs', ascending=False).head(10)
# Print the top 10 apps
print(top_10_apps)
```
